## Cleaning for prices

```python
def extract_prices(html):
    string = str(html)
    splited_string = string.split('>')
    prices=[]
    monedas=[]
    for element in splited_string:
        if '$' in element:
            prices.append(element)
            """ Checar si son MXN o USD"""
            if 'MXN' in element:
                moneda = 'MXN'
            elif 'USD' in element:
                moneda = 'USD'
            else:
                moneda = 'MXN'
            monedas.append(moneda)
    
    for i in range (len(prices)):
        prices[i]=(prices[i].replace(',',''))
        index = prices[i].index('$')
        end = prices[i].index('<')
        try:
            prices[i]=prices[i][index+1:end]
            prices[i]= int(prices[i])
        except:
            # Cortar al inicio del numero
            #index = prices[i].index('$')
            #print(index)
            #prices[i]=prices[i][index+1:]
            #print(prices[i])

            # Cortar al final
            end = prices[i].index(' ')
            #print(end)
            prices[i]=(prices[i][:end])

            # Volver enteros
            prices[i]=int(prices[i])
        
    return(prices, monedas)
```
